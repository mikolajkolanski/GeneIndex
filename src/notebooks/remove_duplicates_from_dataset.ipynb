```python
TRAIN_DS_PATH = 'datasets/full_scans_births_10perc'
TRAIN_DS_PATH2 = 'datasets/full_scans_births_10perc_distil'
NUM_CHOOSEN = 300
```

```python
from src.u1_downloader.images_loader import load_all
from src.u2_image_row_splitter.images_fext_dino import DinoFeatureExtractor

from sklearn.cluster import KMeans 
from sklearn.metrics import pairwise_distances_argmin_min
from sklearn.decomposition import PCA
from sklearn.preprocessing import normalize
import shutil
```

```python
df = load_all(TRAIN_DS_PATH)
```

```python
dfex = DinoFeatureExtractor()
```

```python
feat = dfex.batch_extract_paths(df.image_path)
```

```python
X = normalize(feat)
pca = PCA(n_components=64)
X_pca = pca.fit_transform(X)

km = KMeans(n_clusters=NUM_CHOOSEN)
km.fit(X_pca) 

choosen_idx = pairwise_distances_argmin_min(km.cluster_centers_, X_pca)[0]
mask = df.index.isin(choosen_idx)
df_to_delete = df.iloc[~mask]
```

```python
for _, r in df.iloc[mask].iterrows():
    # print(r)
    print('Copy', r.image_path)
    shutil.copy(r.image_path, TRAIN_DS_PATH2 + '/' + r.image_path.split('/')[-1])
```
